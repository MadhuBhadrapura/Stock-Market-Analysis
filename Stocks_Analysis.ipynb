## 1. Reading the Data from Yahoo finance website directly

```python
import datetime as dt
from dateutil.relativedelta import relativedelta as rd
```

```python
start = dt.datetime.today() - rd(years = 1 )
end = dt.datetime.today() - rd(day=1, days=1)
```

```python
import pandas as pd
import pandas_datareader.data as  pd_data
```

```python
apple = pd_data.DataReader('AAPL', 'yahoo', start, end)
google = pd_data.DataReader('GOOGL', 'yahoo', start, end)
microsoft = pd_data.DataReader('MSFT', 'yahoo', start, end)
amazon = pd_data.DataReader('AMZN', 'yahoo', start, end)
```

## 2. Performing cleaning

```python
#1
apple.isnull()
apple.dropna(inplace = True)
from numpy import NaN
apple.replace({NaN:0.00})
#2
google.isnull()
google.dropna(inplace = True)
from numpy import NaN
google.replace({NaN:0.00})
#3
microsoft.isnull()
microsoft.dropna(inplace = True)
from numpy import NaN
microsoft.replace({NaN:0.00})
#4
amazon.isnull()
amazon.dropna(inplace = True)
from numpy import NaN
amazon.replace({NaN:0.00})
```

## 5. Daily return average of a stock

```python
apple['Daily Return'] = apple['Adj Close'].pct_change()*100
google['Daily Return'] = google['Adj Close'].pct_change()*100
microsoft['Daily Return'] = microsoft['Adj Close'].pct_change()*100
amazon['Daily Return'] = amazon['Adj Close'].pct_change()*100
```

```python
apple['Daily Return'].hist(bins = 100, label = 'Apple', alpha = 0.5, figsize = (15,7))
google['Daily Return'].hist(bins = 100, label = 'Google', alpha = 0.5)
microsoft['Daily Return'].hist(bins = 100, label = 'Microsoft', alpha = 0.5)
amazon['Daily Return'].hist(bins = 100, label = 'Amazon', alpha = 0.5)
plt.legend()
```

## 6. Add a new column â€˜Trendâ€™ whose values are based on the 'Daily Return'

```python
def trend(x):
  if x > -0.5 and x <= 0.5:
    return 'Slight or No change'
  elif x > 0.5 and x <= 1:
    return 'Slight Positive'
  elif x > -1 and x <= -0.5:
    return 'Slight Negative'
  elif x > 1 and x <= 3:
    return 'Positive'
  elif x > -3 and x <= -1:
    return 'Negative'
  elif x > 3 and x <= 7:
    return 'Among top gainers'
  elif x > -7 and x <= -3:
    return 'Among top losers'
  elif x > 7:
    return 'Bull run'
  elif x <= -7:
    return 'Bear drop'

apple['Trend']= apple['Daily Return'].apply(lambda x:trend(x))
google['Trend']= google['Daily Return'].apply(lambda x:trend(x))
microsoft['Trend']= microsoft['Daily Return'].apply(lambda x:trend(x))
amazon['Trend']= amazon['Daily Return'].apply(lambda x:trend(x))
```

## 7. Visualizing trend frequency through a Pie Chart

```python
colors = ["turquoise", "hotpink", "fuchsia",'aquamarine', "teal", "plum", "lightcoral", "purple"]
```

```python
av = apple.Trend.value_counts()
av.values
plt.figure(figsize = (20, 5))
plt.pie(av.values, 
        labels= av.index, autopct= '%2.2f %%',
       textprops= {'color' : 'fuchsia', 'size' : 9, 'weight' : 'bold'},
       wedgeprops= {'edgecolor' : 'gray'},
        colors =colors, shadow = True
       )
plt.title('Trend of Apple stocks', color = 'brown', weight = 'bold', size = 20, pad = 5, loc = 'left')
plt.show()
```
